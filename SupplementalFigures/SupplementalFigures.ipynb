# Supplementary Figure 4: Validation of eSTR analysis using high coverage genotype calls

```python
#- supp fig a (bubbles) b (size of errors), c (eSTR effect sizes) (save in pdfs/high_coverage_validation)
MINSCORE = 0.5
MINCOV = 5
hicov = pd.read_csv(HICOV_GT_FILE, sep="\t")
# Filter to contain things that were called
hicov = hicov[(hicov["andy:allele1"].apply(str) != ".") &
            (hicov["andy:Q"]>=MINSCORE) &
            (hicov["andy:DP"]>=MINCOV)]
hicov = hicov[(hicov["lob:allele1"].apply(str) != ".") &
            (hicov["andy:allele1"].apply(str) != ".")]
hicov["lob:allele1"] = hicov["lob:allele1"].apply(int)
hicov["lob:allele2"] = hicov["lob:allele2"].apply(int)
hicov["andy:allele1"] = hicov["andy:allele1"].apply(int)
hicov["andy:allele2"] = hicov["andy:allele2"].apply(int)
# Get dosage
hicov["lobstr.dosage"] = hicov.apply(lambda x: x["lob:allele1"]+x["lob:allele2"], 1)
hicov["andy.dosage"] = hicov.apply(lambda x: x["andy:allele1"]+x["andy:allele2"], 1)
hicov["correct"] = hicov.apply(lambda x: (x["andy:allele1"]==x["lob:allele1"]) & (x["andy:allele2"]==x["lob:allele2"]), 1)
hicov["locus"] = hicov.apply(lambda x: x["chrom"]+":"+str(x["start"]), 1)

fig = plt.figure()
fig.set_size_inches((15, 5))

# A: bubble plot of comparison
x = hicov.groupby(["lobstr.dosage","andy.dosage"], as_index=False).agg({"sample": len})
x.columns = ["lobstr.dosage","andy.dosage","count"]
ax = fig.add_subplot(131)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
x_notzero = x[~((x["lobstr.dosage"]==0) & (x["andy.dosage"]==0))]
ax.scatter(x["lobstr.dosage"], x["andy.dosage"], color="gray", edgecolor="black",
           s=x["count"].apply(lambda x: x/2), alpha=0.1)
ax.scatter(x_notzero["lobstr.dosage"], x_notzero["andy.dosage"], color="gray", edgecolor="black",
           s=x_notzero["count"].apply(lambda x: x/2))
ax.set_xlim(left=-30, right=30)
ax.set_ylim(bottom=-30, top=30)
ax.set_xlabel("Dosage - 1000G", size=15)
ax.set_ylabel("Dosage - High cov.", size=15)
ax.set_xticklabels(map(int, ax.get_xticks()), size=12)
ax.set_yticklabels(map(int, ax.get_yticks()), size=12);

# B: size of errors
def GetError(allele, gt1, gt2, period):
    # Get closer allele
    dist1 = abs(gt1-allele)
    dist2 = abs(gt2-allele)
    # get error
    if dist1 < dist2:
        return (allele-gt1)/period
    else: return (allele-gt2)/period
hicov["error1"] = hicov.apply(lambda x: GetError(x["lob:allele1"],x["andy:allele1"], x["andy:allele2"], len(x["andy:motif"])), 1)
hicov["error2"] = hicov.apply(lambda x: GetError(x["lob:allele2"],x["andy:allele1"], x["andy:allele2"], len(x["andy:motif"])), 1)

errors = pd.DataFrame({"errors": list(hicov["error1"]) + list(hicov["error2"]), "count": range(hicov.shape[0]*2)})
e=errors[errors["errors"]!=0].groupby("errors", as_index=False).agg({"count":len}).sort("errors")
e=e[e["errors"].apply(abs) <= 5]
e["perc"] = e.apply(lambda x: x["count"]*1.0/sum(e["count"]), 1)
ax = fig.add_subplot(132)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
ax.bar(list(e["errors"]), list(e["perc"]), color="black", align="center")
ax.set_xlabel("Error size (# repeat units)", size=15)
ax.set_ylabel("Percent of errors", size=15)
ax.set_xticklabels(ax.get_xticks(), size=12)
ax.set_yticklabels(ax.get_yticks(), size=12);

# C: Reproducibility of association tests
FDR_CUTOFF = 0.05
hres = pd.read_csv(HICOV_LINREG_RESULTS_FILE)
estr = pd.read_csv(LINREG_RESULTS_FILE)
estr["signif"] = estr.apply(lambda x: x["best_str"] & (x["qval.gene"]<=FDR_CUTOFF), 1)
hits = estr[estr["signif"]]
merged = pd.merge(hits[["chrom","gene","str.start","p.wald","beta"]],
                  hres[["chrom","gene","str.start","p.wald","beta"]], on=["chrom","gene","str.start"])
colors = []
for i in range(merged.shape[0]):
    if merged["beta_x"].values[i]*merged["beta_y"].values[i]>0:
        colors.append("red")
    else: colors.append("gray")
ax = fig.add_subplot(133)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
ax.scatter(merged["beta_x"], merged["beta_y"], color=colors)
ax.set_xlabel("Beta - 1000G", size=15)
ax.set_ylabel("Beta - High cov.", size=15)
ax.axhline(y=0, color="gray", linestyle="dashed")
ax.axvline(x=0, color="gray", linestyle="dashed");

#################
fig.tight_layout()
```

# Supplementary Figure 5: Expression values are moderately reproducible across platforms

```python
# Load Ensembl mapping
ensemble_to_probe = pd.read_table(PROBE_TO_ENSEMBLE, names=["Ensembl","ILMN"])
probe_counts = ensemble_to_probe.groupby("Ensembl").agg({"ILMN": len})
non_unique_ensemble_ids = probe_counts[probe_counts.ILMN > 1].index.values

# Load RNAseq
rnaseq_expr = pd.read_csv(RNASEQ_EXPR_FILE)
rnaseq_expr["Ensembl"] = rnaseq_expr.index.values
probes_to_remove = [probe for probe in rnaseq_expr["Ensembl"] if rnaseq_expr["Ensembl"].tolist().count(probe) > 1]

# Load array and convert to ENS ids
array_expr = pd.read_table(ARRAY_EXPR_FILE)
array_expr["ILMN"] = array_expr["Normalization REF"]
array_expr = pd.merge(array_expr, ensemble_to_probe, on="ILMN")

# Get list of samples in each
array_samples = set([item for item in array_expr.columns.tolist() if item not in ["Ensembl","ILMN","Normalization REF"]])
rnaseq_samples = set([item for item in rnaseq_expr.columns.tolist() if item not in ["Ensembl"]])
samples_in_both = rnaseq_samples.intersection(array_samples)

# Get list of probes in each
array_genes = set(array_expr["Ensembl"])
rnaseq_genes = set(rnaseq_expr["Ensembl"])
genes_in_both = (rnaseq_genes.intersection(array_genes)).difference(list(non_unique_ensemble_ids) + probes_to_remove)

# Get data matrix of gene x sample in same order for each
array_expr = array_expr[array_expr["Ensembl"].apply(lambda x: x in genes_in_both)][["Ensembl"]+list(samples_in_both)]
rnaseq_expr = rnaseq_expr[rnaseq_expr["Ensembl"].apply(lambda x: x in genes_in_both)][["Ensembl"]+list(samples_in_both)]

# Sort
array_expr = array_expr.sort("Ensembl")
rnaseq_expr = rnaseq_expr.sort("Ensembl")

# Get rid of Ensemble
array_expr = array_expr.drop("Ensembl", 1)
rnaseq_expr = rnaseq_expr.drop("Ensembl", 1)
# Get across sample correlation
across_sample_corrs = []
for sample in samples_in_both:
    a = array_expr[sample].apply(math.log10)
    r = rnaseq_expr[sample].apply(lambda x: math.log10(x+1))
    across_sample_corrs.append(spearmanr(a, r)[0])

# Get rank correlation across each gene
across_gene_corrs = []
for i in range(array_expr.shape[0]):
    a = array_expr.iloc[i,:].rank()
    r = rnaseq_expr.iloc[i,:].rank()
    across_gene_corrs.append(spearmanr(a, r)[0])
    
fig = plt.figure()
fig.set_size_inches((10,5))
ax1 = fig.add_subplot(121)
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
ax1.get_xaxis().tick_bottom()
ax1.get_yaxis().tick_left()
ax1.set_xlim(left=0.66, right=0.75)

bins = numpy.arange(0,1,0.01)
n, bins, patches = ax1.hist(across_sample_corrs, bins, color="black", edgecolor="white")
ax1.set_xlabel("Spearman r", size=15);
ax1.set_ylabel("Number of samples", size=15);
ax1.set_xticklabels(ax1.get_xticks(), size=12);
ax1.set_yticklabels(ax1.get_yticks(), size=12);

ax2 = fig.add_subplot(122)
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.get_xaxis().tick_bottom()
ax2.get_yaxis().tick_left()

bins = numpy.arange(0,1,0.1)
n, bins, patches = ax2.hist(across_gene_corrs, bins, color="black", edgecolor="white")
ax2.set_xlabel("Spearman r", size=15);
ax2.set_ylabel("Number of genes", size=15);
ax2.set_xticklabels(ax2.get_xticks(), size=12);
ax2.set_yticklabels(ax2.get_yticks(), size=12);
fig.tight_layout()
```

```python
# Load data
mark_data = {}
for m in MARK_FILES:
    casefile = MARK_FILES[m]["casefile"]
    ctrlfile = MARK_FILES[m]["ctrlfile"]
    mark_data[m] = {}
    mark_data[m]["case"] = pd.read_csv(casefile, sep="\t", names=["chrom","str.start","pval.linreg","effect.linreg","r2","score","signal"])
    mark_data[m]["ctrl"] = pd.read_csv(ctrlfile, sep="\t", names=["chrom","str.start","pval.linreg","effect.linreg","r2","score","signal"])
colors = ["gray","red"]
fig = plt.figure()
fig.set_size_inches((15, 4))
counter = 1
for mark in mark_data.keys():
    pval = ranksums(mark_data[mark]["ctrl"].score.apply(abs), mark_data[mark]["case"].score.apply(abs))[1]
    ax = fig.add_subplot(1, 6, counter)
    counter = counter + 1
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    bp = ax.boxplot([mark_data[mark]["ctrl"].score.apply(abs),
                     mark_data[mark]["case"].score.apply(abs)],
                    sym="", widths=0.75);
    plt.setp(bp['boxes'], color='black', lw=2)
    plt.setp(bp['whiskers'], color='black', linestyle="solid", lw=2)
    plt.setp(bp['medians'], color="white", lw=2)
    # color boxes
    for i in range(2):
        box = bp['boxes'][i]
        boxX = []
        boxY = []
        for j in range(5):
            boxX.append(box.get_xdata()[j])
            boxY.append(box.get_ydata()[j])
        boxCoords = zip(boxX,boxY)
        boxPolygon = Polygon(boxCoords, facecolor=colors[i])
        ax.add_patch(boxPolygon)
    ax.set_title("%s\np=%.5f"%(mark, pval), size=15);
    ax.set_xticklabels(["ctrls","eSTRs"], rotation=90, size=12);
    ax.set_yticklabels(ax.get_yticks(), size=10);
    ax.set_ylabel("Score", size=15);
fig.tight_layout()
```
